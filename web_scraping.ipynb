# World happiness Index
[source](https://en.wikipedia.org/wiki/World_Happiness_Report)

The **World Happiness Report** is an annual publication of the **United Nations**. It contains articles, and rankings of national happiness based on respondent ratings of their own lives, which the report also correlates with various life factors:

* GDP per capita
* Social support
* Healthy life expectancy
* Freedom to make life choices
* Generosity
* Perceptions of corruption


```python
url = 'https://en.wikipedia.org/wiki/World_Happiness_Report'
page = urlopen(url)
```

```python
soup = BeautifulSoup(page)
```

```python
all_tables = soup.find('table', 'wikitable sortable')
```

```python
rank = []
country = []
score = []
gdp_per_capita = []
social_support = []
health_lie_expectancy = []
choice_freedom_index = []
generosity = []
perception_of_corruption = []

for row in all_tables.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) > 0:
        rank.append(cells[0].text.strip())
        country.append(cells[1].text.strip())
        score.append(cells[2].text.strip())
        gdp_per_capita.append(cells[3].text.strip())
        social_support.append(cells[4].text.strip())
        health_lie_expectancy.append(cells[5].text.strip())
        choice_freedom_index.append(cells[6].text.strip())
        generosity.append(cells[7].text.strip())
        perception_of_corruption.append(cells[8].text.strip())
```

```python
data = np.array([rank, country, score, gdp_per_capita, social_support, health_lie_expectancy, choice_freedom_index, generosity, perception_of_corruption]).T
```

```python
df = pd.DataFrame(data, columns=['rank', 'country', 'score', 'gdp_per_capita', 'social_support', 'health_lie_expectancy', 'choice_freedom_index', 'generosity', 'perception_of_corruption'])
```

```python
df.to_csv(data_path + '\indexes\\world_happiness_index.csv', index=False)
```

# Corruption Perceptions Index
[source](https://en.wikipedia.org/wiki/Corruption_Perceptions_Index)

```python
rank = []
country = []
cpi_2019 = []

for row in all_tables[0].find_all('tr'):
    cells = row.find_all('td')
    if len(cells) > 0:
        rank.append(cells[0].text.strip())
        country.append(cells[1].text.strip())
        cpi_2019.append(cells[2].text.strip())
```

# Democracy Index
[source](https://en.wikipedia.org/wiki/Democracy_Index)

The Democracy Index is an index compiled by the Economist Intelligence Unit (EIU), a UK-based company. It intends to measure the state of democracy in 167 countries, of which 166 are sovereign states and 164 are UN member states.

```python
url = 'https://en.wikipedia.org/wiki/Democracy_Index'
page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')
```

```python
all_tables = soup.find_all('table', 'wikitable sortable')
```

```python
by_country_t = all_tables[0]
by_region_t = all_tables[1]
```

```python
rank = []
region = []
num_of_conutry = []
dem_index_2019 = []

for row in by_region_t.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) > 0:
        rank.append(cells[0].text.strip())
        region.append(cells[1].text.strip())
        num_of_conutry.append(cells[2].text.strip())
        dem_index_2019.append(cells[-1].text.strip())
```

```python
data = np.array([rank, region, num_of_conutry, dem_index_2019]).T
```

```python
df = pd.DataFrame(data, columns=['rank', 'region', 'num_of_country', 'democracy_index_2019'])
```

```python
df.to_csv(data_path + '\indexes\democracy_index_by_region_2019.csv', index = False)
```
